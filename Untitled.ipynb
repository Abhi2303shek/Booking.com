```python
import pandas as pd
```

```python
booking = pd.read_csv('booking_hotel.csv', sep=',', encoding='latin1')
booking
```

```python
booking['Hotel Name'] = booking['Hotel Name'].str.replace("?", '', regex=False)
```

```python
booking['Hotel Name'] = booking['Hotel Name'].str.strip('?')
```

```python
booking.columns
```

```python
booking["Rating"] = booking["Rating"].str.extract(r'(\d+\.?\d*)')
booking["Rating"] = pd.to_numeric(booking["Rating"])
```

```python
booking["Number of   "] = booking["Number of   "].str.extract(r'(\d+\.?\d*)')
booking["Number of   "] = pd.to_numeric(booking["Number of   "])
```

```python
#booking["Room    Score"] = booking["Room    Score"].str.extract(r'(\d+\.?\d*)')
booking["Room    Score"] = pd.to_numeric(booking["Room    Score"])
```

```python
booking["Bed Type"] = booking["Bed Type"].str.split(" or ").str[0]
```

```python
booking["room_category"] = (booking["Room Type"]
    .str.extract(r'(Single|Double|Twin|Deluxe|Capsule|Dormitory|Loft)', expand=False)
)
```

```python
booking["has_balcony"] = booking["Room Type"].str.contains("Balcony", case=False, na=False).astype(int)

booking["sea_view"] = booking["Room Type"].str.contains("Sea View", case=False, na=False).astype(int)

booking["shared_bathroom"] = booking["Room Type"].str.contains("Shared Bathroom", case=False, na=False).astype(int)

booking["no_windows"] = booking["Room Type"].str.contains("No Windows", case=False, na=False).astype(int)
```

```python
booking['Room Price (in BDT or any other currency)'].values
```

```python
booking["Room Price (in BDT or any other currency)"] = (
    booking["Room Price (in BDT or any other currency)"]
    .str.replace("\xa0", "", regex=False)   # remove non-breaking space
    .str.replace(",", "", regex=False)      # remove commas
    .str.strip()                            # remove leading/trailing spaces
    .astype(float)                          # or int if no decimals
)
```

```python
booking["bed_count"] = (
    booking["Bed Type"]
    .str.extract(r"(\d+)")
    .fillna(0)
    .astype(int)
)
```

```python
booking["bed_type"] = booking["Bed Type"].str.replace(r"\d+", "", regex=True).str.strip()
```

```python
booking.columns = ['hotel', 'location', 'rating', 'review', 'no_of_reviews', 'room_score',
                  'room_type', 'Bed_TYPE', 'price', 'room_category', 'balcony', 'sea_view',
                  'shared_bathroom', 'no_window', 'bed_count', 'bed_type']
```
